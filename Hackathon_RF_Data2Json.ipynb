```python
import json
import numpy as np
import pandas as pd
import os
import time
import random
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import warnings

warnings.filterwarnings(action='ignore')
```

```python
# Milkcow Data 전처리
# Json 파일 물러와서 DataFrame으로 변환
cow = []
jdir = os.listdir('C:/Users/gjaischool/Desktop/hack/cow/cow/train_jason/')
print(len(jdir))
for k in jdir:

    with open(f'C:/Users/gjaischool/Desktop/hack/cow/cow/train_jason/{k}', 'r', encoding='UTF-8') as f:
        js = json.loads(f.read()) ## json 라이브러리 이용

    col_img = js['ANNOTATION_INFO']

    for i in col_img:
        a = []
        for j in range(39):
            if (j+1) % 3 != 0:
                a.append(i['KEYPOINTS'][j])
        a.append(i['ACTION_NAME'])
        cow.append(a)
    
cow = np.array(cow)
df_cow = pd.DataFrame(cow)
print(df_cow)
```

```python
# Train 위해 특징, 라벨값 분리
mc_X = df_cow.iloc[:,:26]
mc_X
mc_y = df_cow.iloc[:,26:]
mc_y.values
mc_z = df_cow[26]
mc_z.unique()
```

```python
# 훈련용, 테스트용 데이터 행 분리
mc_X_train, mc_X_test, mc_y_train, mc_y_test = train_test_split(mc_X, 

                                                    mc_y, 

                                                    test_size=0.3, 

                                                    shuffle=True, 

                                                    random_state=1004)
```

```python
# 모델 생성
mc_knn = KNeighborsClassifier()
mc_lr = LogisticRegression(max_iter=10000)
mc_svc = SVC()
mc_rf = RandomForestClassifier()
```

```python
# 모델 학습, 예측, 평가
mc_knn.fit(mc_X_train, mc_y_train)
mc_pred_knn = mc_knn.predict(mc_X_test)
print("mc_KNeighborsClassifier : ",mc_knn.score(mc_X_test, mc_y_test))
mc_lr.fit(mc_X_train, mc_y_train)
mc_pred_lr = mc_lr.predict(mc_X_test)
print("mc_LogisticRegression : ",mc_lr.score(mc_X_test, mc_y_test))
mc_svc.fit(mc_X_train, mc_y_train)
mc_pred_svc = mc_svc.predict(mc_X_test)
print("mc_SVC : ",mc_svc.score(mc_X_test, mc_y_test))
mc_rf.fit(mc_X_train, mc_y_train)
mc_pred_rf = mc_rf.predict(mc_X_test)
print("mc_RandomForestClassifier : ",mc_rf.score(mc_X_test, mc_y_test))
```
